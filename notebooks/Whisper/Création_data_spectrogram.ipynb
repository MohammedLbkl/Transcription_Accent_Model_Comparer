```python
import whisper
import pandas as pd
import numpy as np
import torch.nn as nn
import warnings
warnings.filterwarnings("ignore")
```

```python
metadata = pd.read_csv("../../data/processed/Common Voice.csv")
metadata = metadata[["wav_path", "accents"]]
metadata = metadata.dropna(subset=["accents"])
metadata = metadata.reset_index(drop=True) 
metadata
```

```python
model = whisper.load_model("turbo")
```

```python
len(metadata)
```

```python
# Crée une ligne avec 3839 colonnes
row = np.random.rand(3840)
df = pd.DataFrame([row])
df = df.drop(0)
df
```

```python
l=[]
for i in range(len(metadata)):
    audio = whisper.load_audio(f'../../clips.wav/{metadata["wav_path"].iloc[i]}')
    audio = whisper.pad_or_trim(audio)
    mel = whisper.log_mel_spectrogram(audio, n_mels=model.dims.n_mels).to(model.device)
    pool = nn.AvgPool1d(kernel_size=100, stride=100)
    mel_pooled = pool(mel).squeeze(1)
    mel_numpy = mel_pooled.cpu().numpy()
    flat_array = mel_numpy.flatten()
    l.append(flat_array)
```

```python
new_row = pd.DataFrame(l)

df = pd.concat([df, new_row], ignore_index=True)
df
```
