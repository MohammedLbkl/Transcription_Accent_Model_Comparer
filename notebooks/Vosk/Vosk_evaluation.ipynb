```python
import pandas as pd
from vosk import Model, KaldiRecognizer
import wave
import json
from jiwer import wer
import re
import time

import sys
import os
sys.path.append(os.path.abspath("../.."))

from src.audio.audio import Ecoute_audio
from src.models.evaluate_model import vosk_evaluation
from src.models.transcription_model import vosk_transcription
```

```python
metadata = pd.read_csv('../../data/processed/Common Voice.csv')
metadata=metadata[["path","sentence"]]
metadata['wav_path'] = metadata['path'].str.replace('.mp3', '.wav', regex=False)
```

```python
AUDIO_PATH = "../../clips.wav/common_voice_en_40865211.wav"
```

```python
file_path = AUDIO_PATH

Ecoute_audio(file_path)
```

```python
wf =  wave.open(AUDIO_PATH, 'rb')
model = Model(lang="en-us")
rec = KaldiRecognizer(model, wf.getframerate())
rec.SetWords(True)
rec.SetPartialWords(True)

while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        print(rec.Result())
    else:
        print(rec.PartialResult())
        
result = json.loads(rec.Result())
print(result.get("text", ""))
```

```python
wf = wave.open(file_path, 'rb')
model = Model(lang="en-us")
rec = KaldiRecognizer(model, wf.getframerate())
rec.SetWords(True)

full_text = ""

while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        result = json.loads(rec.Result())
    else:
        print(json.loads(rec.PartialResult()))
        
final_result = json.loads(rec.FinalResult())
full_text = final_result.get("text", "")
print(full_text)
```
